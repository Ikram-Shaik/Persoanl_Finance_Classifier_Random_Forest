```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df["Income"] = round(df["Income"])
```

```python
df.head()
```

```python
df.head()
```

```python
df.columns
```

```python
df["Fixed Expenses"] = df["Rent"] + df["Insurance"] + df["Groceries"] + df["Transport"] + df["Eating_Out"] + df["Entertainment"] + df["Utilities"] + df["Healthcare"] + df["Education"] + df["Miscellaneous"]
```

```python
df.head()
```

```python
desired_savings_investment_ratio = 0.90 # 90% of desired savings goes to investment
buffer_investment_ratio = 0.03
```

```python
df['Investment'] = 0.0
```

```python
df['Investment'] = np.maximum(
    0, # Investment should not be negative
    (df['Desired_Savings'] * desired_savings_investment_ratio) +
    (df['Disposable_Income'] * buffer_investment_ratio)
)
```

```python
df['Investment'] = np.minimum(df['Investment'], df['Disposable_Income'])
```

```python
df.loc[df['Disposable_Income'] < 0, 'Investment'] = 0
```

```python
df['Investment'] = df['Investment'].round(2)
df['Loan_Repayment'] = df['Loan_Repayment'].round(2)
```

```python
df.head()
```

```python
df['Savings'] = df['Disposable_Income'] - df['Investment']
```

```python
df.loc[df['Savings'] < 0, 'Savings'] = 0
```

```python
df['Savings'] = df['Savings'].round(2)
```

```python
df.head()
```

```python
df.columns
```

```python
df["Eating_Out"] = round(df["Eating_Out"])
df["Entertainment"] = round(df["Entertainment"])
df["Miscellaneous"] = round(df["Miscellaneous"])
```

```python
df[["Eating_Out","Entertainment","Miscellaneous"]].describe()
```

```python
df['Shopping_Frequency'] = (
    (df['Eating_Out'] / 500) +          # Assume ₹1000 per restaurant visit
    (df['Entertainment'] / 1000) +      # Assume ₹1000 per outing
    (df['Miscellaneous'] / 300)         # Assume ₹500 per small purchase
).round().astype(int)

# Cap at reasonable maximum (e.g., 15 trips/month)
df['Shopping_Frequency'] = df['Shopping_Frequency'].clip(upper=15)
```

```python
df['Shopping_Frequency'].describe()
```

```python
# Corrected approach using row-wise conditions
def get_hobbies(row):
    hobbies = []
    
    # Expense-based hobbies
    if row['Entertainment'] > 3000:
        hobbies.extend(['Travel'])
    if row['Miscellaneous'] > 2000:
        hobbies.extend(['Gaming', 'Electronics'])
    if row['Healthcare'] < 1000:
        hobbies.extend(['Fitness', 'Sports'])
    if row['Entertainment'] < 1000:
        hobbies.extend(['Reading', 'Writing'])
    if row['Groceries'] > row['Eating_Out']:
        hobbies.extend(['Cooking', 'Baking'])
    if row['Miscellaneous'] > 1500:
        hobbies.extend(['Photography'])
    
    # Demographic-based additions
    if row['Occupation'] == 'Student':
        hobbies.extend(['Video Games', 'Social Media'])
    if row['Age'] > 50:
        hobbies.extend(['Gardening', 'Reading'])
    
    # Select 1-2 hobbies randomly
    if len(hobbies) > 0:
        selected = np.random.choice(hobbies, size=min(2, len(hobbies)), replace=False)
        return ', '.join(selected)
    return 'General'

df['Hobbies'] = df.apply(get_hobbies, axis=1)
```

```python
final_df = df[['Income', 'Age','Occupation', 'City_Tier','Loan_Repayment','Fixed Expenses', 'Investment',
       'Savings', 'Shopping_Frequency', 'Hobbies']]
```

```python
final_df.head()
```

```python
final_df.isna().sum()
```

```python
final_df.to_csv("final_dataset.csv",index=False)
```

```python
def determine_class(row):
    # Calculate key ratios
    savings_ratio = row['Savings'] / row['Income']
    investment_ratio = row['Investment'] / row['Income']
    spending_ratio = (row['Fixed Expenses'] + row['Loan_Repayment']) / row['Income']
    
    # Rule 1: Heavy investors are always Savers
    if investment_ratio > 0.25:
        return 'Saver'
    
    # Rule 2: High spending with low savings
    if spending_ratio > 0.7 and savings_ratio < 0.1:
        return 'Spender'
    
    # Rule 3: Frequent shoppers with high spending score
    if row['Shopping_Frequency'] > 10 and spending_ratio > 0.6:
        return 'Spender'
    
    # Rule 4: Consistent savers
    if savings_ratio > 0.2 and investment_ratio > 0.1:
        return 'Saver'
    
    # Rule 5: Retirees with low spending
    if row['Occupation'] == 'Retired' and spending_ratio < 0.5:
        return 'Saver'
    
    # Default to Neutral
    return 'Neutral'

final_df['Class'] = final_df.apply(determine_class, axis=1)
```

```python
final_df.head()
```

```python
final_df.Class.value_counts()
```

```python
def determine_class_v2(row):
    # Financial ratios
    savings_ratio = row['Savings'] / row['Income']
    disposable_income = row['Income'] - row['Fixed Expenses'] - row['Loan_Repayment']
    
    # Hobbies impact
    hobby_factor = 1.0
    if 'Travel' in row['Hobbies']:
        hobby_factor -= 0.2
    if 'Gaming' in row['Hobbies']:
        hobby_factor -= 0.1
    if 'Reading' in row['Hobbies']:
        hobby_factor += 0.15
    
    # Adjusted savings threshold
    effective_savings = savings_ratio * hobby_factor
    
    # Classification rules
    if effective_savings > 0.25:
        return 'Saver'
    elif effective_savings < 0.05 or disposable_income < 0:
        return 'Spender'
    elif row['Shopping_Frequency'] > 8:
        return 'Spender'
    else:
        return 'Neutral'

final_df['Class'] = final_df.apply(determine_class_v2, axis=1)
```
